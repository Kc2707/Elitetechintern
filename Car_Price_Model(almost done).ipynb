```python
def get_brand_name(car_name):
    car_name = car_name.split(' ')[0]
    return car_name.strip()
```

```python
def clean_data(value):
    value = value.split(' ')[0]
    value = value.strip()
    if value == '':
        value = 0
    return float(value)
```
